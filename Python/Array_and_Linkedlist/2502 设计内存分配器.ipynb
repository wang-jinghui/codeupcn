
### 题目描述

给你一个整数 n ，表示下标从 0 开始的内存数组的大小。所有内存单元开始都是空闲的。

请你设计一个具备以下功能的内存分配器：
1. 分配 一块大小为 size 的连续空闲内存单元并赋 id mID 。
2. 释放 给定 id mID 对应的所有内存单元。

注意：
- **多个块可以被分配到同一个 mID **。
- 你必须释放 mID 对应的所有内存单元，即便这些内存单元被分配在不同的块中。

实现 Allocator 类：
- Allocator(int n) 使用一个大小为 n 的内存数组初始化 Allocator 对象。
- int allocate(int size, int mID) 找出大小为 size 个连续空闲内存单元且位于**最左侧**的块，分配并赋 id mID 。返回块的第一个下标。如果不存在这样的块，返回 -1 。
- int freeMemory(int mID) 释放 id mID 对应的所有内存单元。返回释放的内存单元数目。
 

示例：

输入
- ["Allocator", "allocate", "allocate", "allocate", "freeMemory", "allocate", "allocate", "allocate", "freeMemory", "allocate", "freeMemory"]
- [[10], [1, 1], [1, 2], [1, 3], [2], [3, 4], [1, 1], [1, 1], [1], [10, 2], [7]]

输出
-[null, 0, 1, 2, 1, 3, 1, 6, 3, -1, 0]

loc.allocate(3, 4); // 最左侧的块的第一个下标是 3 。内存数组变为 [1, ,3,4,4,4, , , , ]。返回 3 。

提示：

1 <= n, size, mID <= 1000
最多调用 allocate 和 free 方法 1000 次

解题思路：
1. 数据结构选择：
- 使用一个数组 memory 来表示内存单元的状态，数组的大小为 n，初始值为 0，表示所有内存单元都是空闲的。
- 使用一个字典 allocation_map 来记录每个 mID 占用的内存块，键为 mID，值为一个列表，列表中的每个元素是一个元组 (start, size)，表示 mID 占用的内存块的起始位置和大小。
2. 初始化：
- 在初始化时，创建一个大小为 n 的数组 memory，并将所有元素初始化为 0。
- 创建一个空的字典 allocation_map。
3. 分配内存：
- 在 allocate 方法中，遍历 memory 数组，寻找大小为 size 的连续空闲内存块。
- 如果找到这样的内存块，将其标记为已分配（即将对应位置的值设置为 mID），并更新 allocation_map。
- 返回该内存块的第一个下标。
- 如果没有找到这样的内存块，返回 -1。
4.释放内存：
- 在 freeMemory 方法中，根据 mID 查找 allocation_map，获取所有被 mID 占用的内存块。
- 将这些内存块在 memory 数组中标记为空闲（即将对应位置的值设置为 0）。
- 更新 allocation_map，移除 mID 对应的记录。
- 返回释放的内存单元数目。

````python
# out of time
class Allocator:
    def __init__(self, n: int):
        self.n = n
        self.memory = [0] * n 
        self.used = {}

    def allocate(self, size: int, mID: int) -> int:
        for i in range(self.n - size + 1):
            if all(self.memory[i + j] == 0 for j in range(size)):
                for j in range(size):
                    self.memory[i + j] = mID
                if mID not in self.used:
                    self.used[mID] = []
                self.used[mID].append((i, size))
                return i
        return -1
    
    def freeMemory(self, mID: int) -> int:
        if mID not in self.used:
            return 0
        size = sum(size for _, size in self.used[mID])
        for start, size in self.used[mID]:
            for j in range(size):
                self.memory[start + j] = 0
        del self.used[mID]
        return size
````

**加速**：
以下是优化后的思路：
1. 使用区间树（Interval Tree）或段树（Segment Tree）：
- 这些数据结构可以高效地查询和更新区间状态，适合处理连续内存分配的问题。
- 但是实现起来较为复杂，对于简单的场景可能不需要。
2. 使用链表或块列表：
- 维护一个有序的空闲内存块列表，每个块包含起始位置和大小。
- 在分配内存时，查找第一个满足大小要求的空闲块，并将其分割成已分配部分和剩余部分。
- 在释放内存时，将释放的内存块合并到空闲块列表中，并合并相邻的空闲块。
- 这种方法的时间复杂度在最坏情况下为 O(n log n)，因为每次插入和删除操作都需要对列表进行排序。通过这种方式，可以显著提高内存分配和释放的效率。
3. 使用位图（Bitmap）：
- 使用位图来表示内存单元的状态，可以高效地进行位操作。
- 但是位图的实现和操作相对复杂。

下面是使用块列表实现：

````python
class Allocator:
    def __init__(self, n: int):
        self.n = n
        self.free_blocks = [(0, n)]
        self.used = {}
    
    def allocate(self, size: int, mID: int) -> int:
        for i, (start, block_size) in enumerate(self.free_blocks):
            if block_size >= size:
                self.free_blocks.pop(i)
                if block_size > size:
                    # append or insert
                    self.free_blocks.insert(i, (start + size, block_size - size))
                if mID not in self.used:
                    self.used[mID] = []
                self.used[mID].append((start, size))
                self.free_blocks.sort()
                return start
        return -1 
    
    def freeMemory(self, mID: int) -> int:
        if mID not in self.used:
            return 0
        free_size = 0
        for start, block_size in self.used[mID]:
            self.free_blocks.append((start, block_size))
            free_size += block_size
        del self.used[mID]
        self.free_blocks.sort()
        self._merge_free_blocks()
        return free_size
    
    def _merge_free_blocks(self):
        i = 0
        while i < len(self.free_blocks) - 1:
            start1, size1 = self.free_blocks[i]
            start2, size2 = self.free_blocks[i + 1]
            if start1 + size1 == start2:
                self.free_blocks[i] = (start1, size1 + size2)
                self.free_blocks.pop(i + 1) # del self.free_blocks[i + 1]
            else:
                i += 1
````
